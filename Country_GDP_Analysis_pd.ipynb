```python
'''
Distplots are used to do univariate analysis on a continious/numerical data.

Distplot combines 3 things in one graph
1. Histogram -> bars/column in the graph (showing the distributions of values)
2. Kernel Density Estimation (KDE) -> line in the graph
3. Rug plot


# using seaborn read dataset using
titanic = sns.load_dataset('titanic')
'''

'''
In Below graph line is KDE(Kernel Density Estimation) showing the probability density of the data on y axis and 
the histogram is showing the distribution of the data on x axis
'''


# distributions 
vis1 = sns.distplot(df['InternetUsers']) # Univariate analysis - distribution plot
# plt.show(vis1) # it will also work 
```

```python
'''
Here the kde is not shown in the graph 
Histogram is counting the frequency of the data in the given bins
'''
# plotting only the histogram
# vis1 = sns.distplot(df['InternetUsers'], bins=10, kde=False) # Seaborn automatically take comfortable bins
vis1 = sns.distplot(df['InternetUsers'], kde=False)
```

```python
'''


'''
# plotting only the kde
vis1 = sns.distplot(df['InternetUsers'], hist=False, color='green')
```

```python
'''
The only difference between histogram and rug plot is that histogram is showing the column on the y axis and rug plot is showing the column on the x axis
'''

# plotting only a rug plot
vis1 = sns.distplot(df['InternetUsers'], hist=False, kde = False, rug=True)
```

```python
# displot -> distribution plot
vis2 = sns.displot(df['InternetUsers']) # modified version of distplot
```

```python
vis3 = sns.distplot(df['InternetUsers'], bins=10) # bins means 10 columns is created in graph
```
